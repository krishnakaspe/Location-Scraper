# LOCATION FINDER 
## Takes input as userid finds their 11 posts(allowed by instagram) addresses which include
|Categories|Description
|--|--|
|sc[i]|user_id
|l_id|locationid
|slug|location name
|lat|Latitude
|lng|Longitude
|s_address|Street address
|zip_code|ex:500085
|city_name|city name
|region_name|regionname
|c_code|country code



### Logic :
###### Phase I :Find location id and slug from source code of user_id t
###### Phase II:Then make a new request which include locationid and slug in the url then scrape all the address ,geolocation  from the new source code 

```python
import requests,re
from bs4 import BeautifulSoup as bs
import requests
import csv
import re
import json
import urllib.request
from bs4 import BeautifulSoup as bs
import random,time
#from exceptions import AssertionError

user_agent_list = []
useragent=open('Useragent.txt','r')
for ug in useragent:
    user_agent_list.append(ug.strip())
#random.shuffle(user_agent_list)
sc=[]
count=0
rand=0
f=open('userids2.csv','r',encoding='utf-8')
for line in f:
    sc.append(line.strip())
for i in range (0,len(sc)):
    #user_agent=user_agent_list[random.randint(0,4656)]
    if rand==count:
        count=0        
        rand=random.randint(5,10)
        user_agent=user_agent_list[random.randint(0,4656)]
    try:
        r = requests.get("https://www.instagram.com/"+sc[i],headers={'User-Agent':user_agent })
    except OSError:
        print('Error at',i+1)
        user_agent=user_agent_list[random.randint(0,4656)]
        r = requests.get("https://www.instagram.com/"+sc[i],headers={'User-Agent':user_agent })
    if r.status_code == 200 or r.status_code == 429:
        soup = bs(r.text)
    else:
        print('Broken')
        continue
    x_new=[];loc=[];slg=[];slug='';l_id=''
    slugdata=soup.find_all('script', attrs={'type': 'text/javascript'})
    y1=str(slugdata[3])
    #print(y)
    x1=re.findall(r'"([^"]*)"',y1)
    indices=[i for i,x_new in enumerate(x1)if 'location' == x_new]

    for num in range(0,len(indices)):
        #print(x[indices[num]+2])
        if(x1[indices[num]+2]!='media_preview'):
            l_id=x1[indices[num]+2]
            #print(l_id)
            slug=x1[indices[num]+7]
            if slug not in slg:
                #print('HH')
                slg.append(slug) 
            else:
                #print('uu')
                continue
        else:
            continue

        if rand==count:
            count=0        
            rand=random.randint(5,10)
            user_agent=user_agent_list[random.randint(0,4656)]
        try:
            r = requests.get("https://www.instagram.com/explore/locations/"+l_id+'/'+slug,headers={'User-Agent':user_agent })
        except OSError:
            print('Error at',i+1)
            user_agent=user_agent_list[random.randint(0,4656)]
            r = requests.get("https://www.instagram.com/explore/locations/"+l_id+'/'+slug,headers={'User-Agent':user_agent })
        if r.status_code == 200 or r.status_code == 429:
             soup = bs(r.text)
        else:
            print('Broken',r.status_code)
            continue

        bs_data = bs(r.text,'html.parser')
        x=[]
        x=bs_data.find_all('script', attrs={'type': 'text/javascript'})
        y=str(x[3])
        lk=[]

        lk=re.findall(r'\b\d+.\d+\b',str(x[3]))
        if len(lk)!=0:
            #ghu=[next(lk).group(0) for _ in range(3)]
            lat=lk[1]
            lng=lk[2]
        
            
        else:
            print(sc[i]+'search this guys data again')
            time.sleep(10)
            continue
        email=re.findall(r'\"(.+?)\"',y)
        if 'name' in email:
            name=email[email.index('name')+1]
            if name=='\\':
                name='NA'
            else:
                name=name.replace('\\','')
        if 'street_address\\' in email:
            s_address=email[email.index('street_address\\')+1]
            if s_address=='\\':
                s_address='NA'
            else:
                s_address=s_address.replace('\\','')
        else:
            s_address='NA'
        if 'zip_code\\' in email:
            zip_code=email[email.index('zip_code\\')+1]
            if zip_code=='\\':
                zip_code='NA'
            else:
                zip_code=zip_code.replace('\\','')
        else:
            zip_code='NA'
        if 'city_name\\' in email:
            city_name=email[email.index('city_name\\')+1]
            if city_name=='\\':
                city_name='NA'
            else:
                city_name=city_name.replace('\\','')
        else:
            city_name='NA'
        if 'region_name\\' in email:
            region_name=email[email.index('region_name\\')+1]
            if region_name=='\\':
                region_name='NA'
            else:
                region_name=region_name.replace('\\','')
        else:
            region_name='NA'
        if 'country_code\\' in email:
            c_code=email[email.index('country_code\\')+1]
            if c_code=='\\':
                c_code='NA'
            else:
                c_code=c_code.replace('\\','')
        else:
            c_code='NA'
        row=[sc[i],l_id,slug,name,lat,lng,s_address,zip_code,city_name,region_name,c_code]
        print(row)
        count+=1
        with open('locationdetails.csv', 'a',encoding='utf-8') as csvFile:
            writer = csv.writer(csvFile)
            writer.writerow(row)
        csvFile.close() 
                                             
                                                  
```
